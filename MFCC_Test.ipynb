```python
# MFCC-Test
#
# Created By ravissement On 2021-11-05.

import os
import numpy as np
import pandas as pd
from tqdm import tqdm

import librosa
import librosa.display
import matplotlib.pyplot as plt

path = 'd:/ravissement/pythonData/train16K/omy_16k.wav'

y, sr = librosa.load(path,sr=None) # y: Audio data(array), sr: sampling rate

# librosa.feature.melspectrogram(y=y, sr=8000)
D = np.abs(librosa.stft(y))**2  # Short Time Fourier Transform 
print(D)

#S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=sr/2) 
S = librosa.feature.melspectrogram(S=D, sr=sr)
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=sr/2, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Mel-frequency spectrogram')
# plt.savefig(f'D:/.jpg')
plt.show()
```
